```
import csv
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
```

```
FILENAME = 'student-mat.csv'
students = []

with open(FILENAME, newline="") as file:
  reader = csv.DictReader(file)
  for row in reader:
    age = int(row['age'])
    sex = row['sex']
    pstatus = row['Pstatus']
    study_time = int(row['studytime'])
    activities = row['activities']
    father_job = row['Fjob']
    mother_job = row['Mjob']
    dalc = int(row['Dalc'])
    walc = int(row['Walc'])
    absences = int(row['absences'])

    students.append({'age': age,
                    'sex': sex,
                    'pstatus': pstatus,
                    'study_time': study_time,
                    'activities': activities,
                    'father_job': father_job,
                    'mother_job': mother_job,
                    'dalc': dalc,
                    'walc': walc,
                    'absences': absences})
```

Statistic information:

```
rate = [student['dalc'] for student in students]

from statistics import *
print(mean(rate))
print(median(rate))
plt.boxplot(rate, vert=False)
```

Compare the alcohol comsumption by sex during the weekday

```
def weekday_by_sex():
  rate_male = [student['dalc'] for student in students if student['sex'] == 'M']
  rate_female = [student['dalc'] for student in students if student['sex'] == 'F']
  counter_male = Counter(rate_male)
  counter_female = Counter(rate_female)

  num_of_rate_male = []
  num_of_rate_female = []
  
  for key in range(1,6):
    num_of_rate_male.append(counter_male.get(key))
    num_of_rate_female.append(counter_female.get(key))

  rates = np.arange(1,6) 
  plt.bar(rates - .15, num_of_rate_male, width=0.35, color='blue', label='Male')
  plt.bar(rates + .2, num_of_rate_female, width=0.35, color='pink', label='Female')
  plt.legend()
  plt.xlabel("Alcohol consumption (1 : very low - 5 : very high)")
  plt.ylabel("Numbers of students")
  plt.title("Alcohol consumption between male and female consume alcohol during weekday.")
  plt.show()

weekday_by_sex()
```

Compare the alcohol comsumption by attending extra-curricular

```
def by_extra_curricular():
  rate_yes = [student['dalc'] for student in students if student['activities'] == 'yes']
  rate_no = [student['dalc'] for student in students if student['activities'] == 'no']
  counter_yes = Counter(rate_yes)
  counter_no = Counter(rate_no)

  num_of_rate_yes = []
  num_of_rate_no = []
  
  for key in range(1,6):
    num_of_rate_yes.append(counter_yes.get(key))
    num_of_rate_no.append(counter_no.get(key))

  rates = np.arange(1,6) 
  plt.plot(rates, num_of_rate_yes, linewidth=2, color='blue', label='Attend')
  plt.plot(rates, num_of_rate_no, linewidth=2, color='red', label='No-attend')
  plt.legend()
  plt.xlabel("Alcohol consumption (1 : very low - 5 : very high)")
  plt.ylabel("Numbers of students")
  plt.title("Alcohol consumption between students attend extra-curricular activities.")
  plt.show()

extra_curricular()
```

Compare the alcohol comsumption by parent's cohabitation status

```
def by_parent_status():
  rate_together = [student['dalc'] for student in students if student['pstatus'] == 'T']
  rate_apart = [student['dalc'] for student in students if student['pstatus'] == 'A']
  counter_together = Counter(rate_together)
  counter_apart = Counter(rate_apart)
  
  num_of_rate_together = []
  num_of_rate_apart = []
  
  for key in range(1,6):
    num_of_rate_together.append(counter_together.get(key))
    num_of_rate_apart.append(counter_apart.get(key))

  rates = np.arange(1,6) 
  plt.plot(rates, num_of_rate_together, linewidth=2, color='skyblue', label='Together')
  plt.plot(rates, num_of_rate_apart, linewidth=2, color='olive', label='Apart', linestyle='dashed')
  plt.legend()
  plt.xlabel("Alcohol consumption (1 : very low - 5 : very high)")
  plt.ylabel("Numbers of students")
  plt.title("Alcohol consumption between students by parent's cohabitation status.")
  plt.show()

by_parent_status()
```

Compare the alcohol comsumption by mother's jobs


```
def by_mother_job():
  teacher = [student['dalc'] for student in students if student['mother_job'] == 'teacher']
  health = [student['dalc'] for student in students if student['mother_job'] == 'health']
  services = [student['dalc'] for student in students if student['mother_job'] == 'services']
  home = [student['dalc'] for student in students if student['mother_job'] == 'at_home']
  other = [student['dalc'] for student in students if student['mother_job'] == 'other']

  counter_teacher = Counter(teacher)
  counter_health = Counter(health)
  counter_services = Counter(services)
  counter_home = Counter(home)
  counter_other = Counter(other)
  
  
  num_of_rate_teacher = []
  num_of_rate_health = []
  num_of_rate_services = []
  num_of_rate_home = []
  num_of_rate_other = []
  
  
  for key in range(1,6):
    num_of_rate_teacher.append(counter_teacher.get(key))
    num_of_rate_health.append(counter_health.get(key))
    num_of_rate_services.append(counter_services.get(key))
    num_of_rate_home.append(counter_home.get(key))
    num_of_rate_other.append(counter_other.get(key))
  
  num_of_rate_teacher = [0 if x==None else x for x in num_of_rate_teacher] 
  num_of_rate_health = [0 if x==None else x for x in num_of_rate_health] 
  num_of_rate_services = [0 if x==None else x for x in num_of_rate_services] 
  num_of_rate_home = [0 if x==None else x for x in num_of_rate_home] 
  num_of_rate_other = [0 if x==None else x for x in num_of_rate_other] 
 
  rates = np.arange(1,6) 
  plt.plot(rates, num_of_rate_teacher, color='blue', label='Teacher')
  plt.plot(rates, num_of_rate_health, color='red', label='Health')
  plt.plot(rates, num_of_rate_services, color='green', label='Services')
  plt.plot(rates, num_of_rate_home, color='purple', label='At home')
  plt.plot(rates, num_of_rate_other, color='orange', label='Other')


  plt.legend()
  plt.xlabel("Alcohol consumption (1 : very low - 5 : very high)")
  plt.ylabel("Numbers of students")
  plt.title("Alcohol consumption between students by mother's jobs.")
  plt.show()

by_mother_job()
```
